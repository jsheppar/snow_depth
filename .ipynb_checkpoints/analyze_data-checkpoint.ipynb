## Parse Data with Sliderule

```python
# region of interest (1)
belknap_roi = [ {"lon": -122.12465, "lat": 44.12274},
               {"lon": -121.75132, "lat": 44.12274},
               {"lon": -121.75132, "lat": 44.50082},
               {"lon": -122.12465, "lat": 44.50082},
               {"lon": -122.12465, "lat": 44.12274} ]
```

```python
granules = icesat2.cmr(polygon=belknap_roi, time_start="2021-01-01", time_end="2021-05-01", short_name='ATL03')
print(granules)
#resource = granules[0]  #0, 4 have data in this date range
#print(resource)
```

```python
# initialize (2)
icesat2.init("icesat2sliderule.org", verbose=True)
```

```python
# processing parameters (3)
parms = {
    "srt": 0,  #allows ATL08 photons to be used
    "cnf": 0,  #no native filtering, allows ATL08 to be used
    "atl08_class": "atl08_ground",
    "ats": 20.0,
    "cnt": 10,
    "len": 40.0,  #length of each extent, try 40, 60, 100
    "res": 20.0,  #distance between subsequent extent centers, try 20, 30, 50
    "maxi": 6
}
```

```python
for filename in granules:
    gdf = icesat2.atl06(parms, filename)
    if len(gdf) > 0:
        gdf['lon'] = gdf['geometry'].x
        gdf['lat'] = gdf['geometry'].y
        up_gdf = gdf[(gdf['lon'] > -122.12465) & (gdf['lon'] < -121.75132) & (gdf['lat'] > 44.12274) & (gdf['lat'] < 44.50082)]
        if len(up_gdf) > 0:
            (up_gdf['x'], up_gdf['y']) = fn.coords_trans(up_gdf['lon'], up_gdf['lat'])
            bare_earth_values, icesat_values = fn.compare_heights_v2(up_gdf, bare_earth_belknap)

            array1 = np.array(bare_earth_values)
            array2 = np.array(icesat_values)
            subtracted_array = np.subtract(array1, array2)
            subtracted = list(subtracted_array)

            print('Mean: ', np.mean(subtracted))
            print('Median: ', np.median(subtracted))
            print('Standard Deviation: ', np.std(subtracted))
```

## Analyze ICESat elevation data compared to bare earth model

```python
bare_earth_values, icesat_values = fn.compare_heights_v2(up_gdf, bare_earth_belknap)
```

```python
plt.figure(figsize=(8,8), constrained_layout=True)
plt.hist2d(bare_earth_values, icesat_values, bins=100, cmap=plt.cm.Reds)
plt.colorbar()

#create x=y line
lims = [
    np.min([bare_earth_values, icesat_values]),  # min of both axes
    np.max([bare_earth_values, icesat_values]),  # max of both axes
]
# plot both limits against eachother
plt.plot(lims, lims, 'r-')

plt.title("Elevation Histogram Between IceSAT2 snow-off and Bare Earth Model")
plt.xlabel('Bare Earth Elevation [m]')
plt.ylabel('IceSAT Elevation [m]')
```

```python
array1 = np.array(bare_earth_values)
array2 = np.array(icesat_values)
subtracted_array = np.subtract(array1, array2)
subtracted = list(subtracted_array)
```

```python
print('Mean: ', np.mean(subtracted))
print('Median: ', np.median(subtracted))
print('Standard Deviation: ', np.std(subtracted))
```
