```python
from sliderule import icesat2
```

```python
# region of interest (1)
belknap_roi = [ {"lon": -122.12465, "lat": 44.12274},
               {"lon": -121.75132, "lat": 44.12274},
               {"lon": -121.75132, "lat": 44.50082},
               {"lon": -122.12465, "lat": 44.50082},
               {"lon": -122.12465, "lat": 44.12274} ]
```

```python
granules = icesat2.cmr(polygon=belknap_roi, time_start="2020-08-01", time_end="2020-10-30", short_name='ATL03')
print(granules[0])
resource = granules[0]
print(resource)
```

```python
# initialize (2)
icesat2.init("icesat2sliderule.org", verbose=True)
```

```python
# processing parameters (3)
parms = {
    "srt": 0,  #allows ATL08 photons to be used
    "cnf": 0,  #no native filtering, allows ATL08 to be used
    "atl08_class": "atl08_ground",
    "ats": 20.0,
    "cnt": 10,
    "len": 40.0,  #length of each extent
    "res": 20.0,  #distance between subsequent extent centers
    "maxi": 6
}
```

```python
# make request (4)
gdf = icesat2.atl06(parms, resource)
gdf
```

```python
gdf['lon'] = gdf['geometry'].x
gdf['lat'] = gdf['geometry'].y
gdf
```

```python
up_gdf = gdf[(gdf['lon'] > -122.12465) & (gdf['lon'] < -121.75132) & (gdf['lat'] > 44.12274) & (gdf['lat'] < 44.50082)]
```
