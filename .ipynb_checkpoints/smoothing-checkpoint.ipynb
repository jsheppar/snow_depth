```python
from skimage import filters
import numpy as np
import xarray as xr
import rasterio
import rioxarray
import os
import os.path
import matplotlib.pyplot as plt
```

```python
if "DOGAMI" in os.getcwd():
    print("Directory not changed.")
    print(os.getcwd())
else:
    os.chdir("./DOGAMI")
    print("Directory changed.")
    print(os.getcwd())
```

```python
dem_og = rioxarray.open_rasterio("corrected_BE_f32.tif")
dem_og = dem_og.squeeze(dim="band", drop=True)  #flatten
dem_og
```

```python
s = 2  #standard deviation to smooth to
w = 25  #define your smoothing window width
t = (((w - 1)/2)-0.5)/s  #truncation term, based on s and w
dem_gaus = filters.gaussian(dem_og, sigma=s, truncate=t, preserve_range=True)
```

```python
vmin = np.min(np.where(dem_gaus > -1e10)) # to ignore nodata values when plotting

im = plt.imshow(dem_gaus, vmin=vmin)
cbar = plt.colorbar(im, shrink=0.8)
cbar.set_label('Elevation [m]', rotation=270, fontsize=14) #colorbar label is a little wonky. fix?
plt.show()
```

```python
da_smooth = xr.DataArray(data=dem_gaus, coords=[dem_og.y,dem_og.x])
```
